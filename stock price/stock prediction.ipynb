```python
train = data[data['Date'] < '2015-01-01']
train
```

```python
test = data[data['Date'] > '2014-12-31']
test
```

```python
# Can also use re  
train1=train.iloc[:,2:27]
train1.replace('[^a-zA-Z]',' ',regex=True,inplace=True)
```

```python
train1
```

```python
#renaming columns names for ease of access
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
train1.columns=new_index
train1.head(2)
```

```python
# Convertin the headline into lower case
for index in new_index:
    train1[index]=train1[index].str.lower()
train1.head(1)       
```

```python
headlines=[]
for i in range(len(train1.index)):
    headlines.append(' '.join(str(x) for x in train1.iloc[i,0:25]))
```

```python
headlines[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#implement BAG of words
CountVector= CountVectorizer(ngram_range=(2,2))
traindataset=CountVector.fit_transform(headlines)
```

```python
traindataset
```

```python
#implement RandomForestClassifier
model=RandomForestClassifier(n_estimators=200,criterion='entropy')
model.fit(traindataset,train['Label'])
```

```python
#prediect for the dataset
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset= CountVector.transform(test_transform)
predictions=model.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
